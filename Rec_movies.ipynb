```python
import pandas as pd
```

```python
movies = pd.read_csv('TMDB_10000_Popular_Movies.csv')
movies.head(10)
```

```python
movies.describe()
```

```python
movies.info()
```

```python
movies = movies[['TMDb_Id','Title','Overview','Genres']]
```

```python
movies.head()
```

```python
movies['Line']=movies['Overview']+movies['Genres']
```

```python
movies.head()
```

```python
movies = movies.drop(columns=['Overview','Genres'])
```

```python
movies
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words='english')
vector_data = cv.fit_transform(movies['Line'].values.astype('U')).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi = cosine_similarity(vector_data)
simi
```

```python
def top_movies(movie):
    idx = movies[movies['Title']==movie].index[0]
    mov = sorted(list(enumerate(simi[idx])),reverse=True, key=lambda vector_data:vector_data[1])
    for i in mov[:5]:
        print(movies.iloc[i[0]].Title)
```

```python
top_movies("Iron Man")
```

```python
import pickle
pickle.dump(movies,open('rec_movies.pkl','wb'))
pickle.dump(simi,open('simi.pkl','wb'))
```
